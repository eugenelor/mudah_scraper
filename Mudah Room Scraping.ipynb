```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
mudah = requests.get("https://www.mudah.my/Selangor/Rooms-2100/master-for-rent?o=1&q=master&so=1&th=1")

mudah_content = mudah.content

mudah_soup = bs(mudah_content, "html.parser")

property_lists=[]

baseurl = "https://www.mudah.my/Selangor/Rooms-2100/master-for-rent?o="
for page in range (1,22):
    print(baseurl+str(page)+"&q=master&so=1&th=1")
    web = requests.get(baseurl+str(page)+"&q=master&so=1&th=1")
    mudah_content = web.content
    mudahsoup = bs(mudah_content, "html.parser")
    allelements = mudahsoup.find_all("div",{"class":"listing_params_container"})

    for i in allelements:
        d = {}
        d["Post Date"] = i.find("div", {"title": "Date & Location"}).text.replace("\n","").replace(",","").strip()[0:16]
#         Grab Locality
        d["Locality"] = i.find("div", {"class":"area"}).text.strip()
#         Grab listing description
        d["Desc"] = i.find("h2", {"class":"list_title"}).text.replace("\n","")
#         Grab listing price
        d["Price"] = i.find("div", {"class": "ads_price"}).text.replace("\n","").strip()
        
#         Grab Property type
        try:
            d["Property_Type"] = i.find("div", {"class":"bottom_info bottom_label rooms"}).text.strip()
        except:
            d["Property_Type"] = None

#       Grab listing house size
        try:
            d["Area"] = i.find("div", {"class":"bottom_info bottom_label size apartmentAndLandAndRoomsAndNew-Properties-size"}).text.strip()
        except:
            d["Area"] = None

#         Grab listing prefered gender. if statements written to grab other three variations of classes
        gCheck = i.find("div", {"class":"bottom_info bottom_label roommate-gender threeValGender"})    
        if gCheck is None:
            gCheck = i.find("div", {"class":"bottom_info bottom_label roommate-gender twoValGender"})
            if gCheck is None:
                gCheck = i.find("div", {"class":"bottom_info bottom_label roommate-gender oneValGender"})
                if gCheck is None:
                    d["Gender"] = "Can't find, sorry bossku"
                else:
                    d["Gender"] = i.find("div", {"class":"bottom_info bottom_label roommate-gender oneValGender"}).text.strip()
            else:
                d["Gender"] = i.find("div", {"class":"bottom_info bottom_label roommate-gender twoValGender"}).text.strip()
        else:
            d["Gender"] = i.find("div", {"class":"bottom_info bottom_label roommate-gender threeValGender"}).text.strip()
        
#         Grab listing url
        try:
            d["Url"] = i.find('a', href=True)["href"]
        except:
            d["Url"] = None
        
        property_lists.append(d)
```

```python
property_lists
```

```python
MudahData = pd.DataFrame(property_lists)
```

```python
MudahData = MudahData[['Post Date', 'Locality', 'Desc', 'Price', 'Gender','Area','Property_Type', 'Url']]
MudahData.to_csv("MudahData.csv")
```
